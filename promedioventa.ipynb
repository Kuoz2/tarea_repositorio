# Tu código aquí:

```python
ventas={"Norte": 4500_000, "Sur": 3200000, "Centro": 6100_000, "Oriente": 2800000 }
for region, venta in ventas.items():
    print(f"{region}: ${venta:,}")
vacio = sum([i for i in ventas.values()])
print(len(ventas))
promedio = vacio/len(ventas)
print(promedio)
print(vacio)

```
