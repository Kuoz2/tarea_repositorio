###  Mini-ejercicio 3 — Clasificador de temperatura



Dado un valor de temperatura (en °C), clasifícalo como:

- < 0°C → `"Bajo cero "`

- 0–15°C → `"Frío "`  

- 16–25°C → `"Templado "`

- 26–35°C → `"Caluroso "`

- > 35°C → `"Extremo "`



Además, imprime si es adecuado para practicar deporte al aire libre (10–28°C).

```python


temp= int(input("Ingrese la temperatura: "))
if temp <= 0: #verdadero y verdadero o false o false
    print("Bajo cero")
elif temp <= 15:
    print("Frio")
elif temp <= 15 or temp < 25:
    print("Templado")
elif temp >= 25 or temp < 35:
    print("Caluroso")
else:
    print("Extremo")
```
