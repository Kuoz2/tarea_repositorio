##Tarea1

```python
def saluda_todos(nom, ed):
    return print(f"Hola, {nom} tu edad es {ed} ")

nombre = input("Ingrese su nombre: ")
edad = int(input("Ingrese su edad: "))

saluda_todos(nombre, edad)
```

##Tarea2

```python
nombre_completo = "David Palta"
minuscula = nombre_completo.lower()
mayuscula = nombre_completo.upper()
print(f"Tu nombre en minuscula: {minuscula}")
print(f"Tu nombre en mayuscula: {mayuscula}")

nombre_cambio = nombre_completo.replace("David", "Favio")
print(f"Nombre completo {nombre_cambio}")
```

##Tarea3

```python
entrada = input("Ingresa una lista de nombres separados por comas: ")
nombres = [nombre.strip() for nombre in entrada.split(",")]

print("\nLista de nombres:")
for nombre in nombres:
    print(nombre)
```
