#Ejercicio 2

```python
def precio_porcentaje(preci_base, pordescuento):
    iva = 0.19
    monto_descuento = preci_base * pordescuento
    precio_des_cont = preci_base - monto_descuento
    # CORRECCIÓN: Usamos + para sumar el 100% del precio más el 19% de IVA
    preci_iva = precio_des_cont * (1 + iva)

    return monto_descuento, precio_des_cont, preci_iva

# CORRECCIÓN: int() envuelve al input para convertir la entrada a número
numero = int(input("Ingrese un precio: "))
# CORRECCIÓN: float() porque el descuento tiene decimales (0.30)
descu = float(input("Ingrese el descuento (ejemplo 0.30): "))

# Ahora enviamos números a la función, no texto
dec, net, total = precio_porcentaje(numero, descu)

print(f"Monto del descuento:     ${dec:,.0f}")
print(f"Precio final (neto):     ${net:,.0f}")
print(f"Precio final con IVA:    ${total:,.0f}")
```
