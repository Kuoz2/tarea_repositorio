## Filtering legendaries

```python
filter = pokedex['Legendary'] == True
legendaries = pokedex[filter]
legendaries
```

## First ten

```python
columnas_requeridas = ['Name', 'Type 1', 'HP', 'Attack', 'Defense', 'Speed']
result = pokedex[columnas_requeridas].head(10)
result
# resultado = pokedex.loc[:9, columnas_requeridas]
# resultado
```

## Powerful Pokemons

```python
filter1 = pokedex['Attack'] > 100
filter2 = pokedex['Defense'] > 100
filter3 = pokedex['Sp. Atk'] > 100
powerful = pokedex[filter1 & filter2 & filter3]
powerful.shape
```
