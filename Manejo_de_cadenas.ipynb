#Ejercicio 2: Manipulación de Cadenas



```python
#Pedir nombre completo
nombre = input("Ingrese su nombre: ")
apellido = input("Ingresa tus apelidos: ")
nombreMin=nombre.lower()
apellidoMin=apellido.lower()
nombreMay=nombre.upper()
apellidoMay=apellido.upper()
nombreCambiado = nombre.replace(nombre, "Alverto")

print(f"Nombre completo: {nombre} {apellido}")
print(f"Nombre en minusculas: {nombreMin}")
print(f"Apellido en minusculas: {apellidoMin}")
print(f"Nombre en mayusculas: {nombreMay}")
print(f"Apellido en mayusculas: {apellidoMay}")
print(f"Nombre cambiado: {nombreCambiado} {apellido}")
```

#Ejercicio 3: Lista de Nombres



```python
lista = []
print("Ingrese nombres de a uno")
print("No coloque nada o escriba salir para mostrar resultado de nombres")
while True:
  datos = str(input("Ingrese nombres por nombre: "))
  if datos.lower() == "salir" or datos == "":
      break

    #SACAMOS LA , Y LUEGO LO PASAMOS A LA LISTA
  lista = [n.strip() for n in datos.split(",")]

  print("Nombres procesados")
  for nombre in lista:
    print(nombre)
  print("-" * 30)
```
